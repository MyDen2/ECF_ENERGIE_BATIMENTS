# Etape 4.3 : Dashboard exécutif

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

FIG_DIR = "../output/figures"
os.makedirs(FIG_DIR, exist_ok=True)

df = pd.read_parquet("../output/consommations_enrichies.parquet")
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df = df.dropna(subset=["timestamp", "conso_clean", "type_energie"])

# utile pour les agrégations
df["date"] = df["timestamp"].dt.date
df["month"] = df["timestamp"].dt.to_period("M").dt.to_timestamp()
```

Créer un dashboard multi-panneaux (2x3) synthétisant les résultats :

- Panel 1 : Evolution de la consommation totale (6 derniers mois)
- Panel 2 : Top 10 bâtiments les plus énergivores (bar horizontal)
- Panel 3 : Répartition des coûts par type d'énergie (pie chart)
- Panel 4 : Consommation moyenne par classe DPE avec écart-type
- Panel 5 : Carte de chaleur par commune
- Panel 6 : Economies potentielles par amélioration du DPE

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
(ax1, ax2, ax3, ax4, ax5, ax6) = axes.flatten()

# Panel 1 

max_date = df["timestamp"].max()
start_6m = max_date - pd.DateOffset(months=6)

df_6m = df[df["timestamp"] >= start_6m].copy()
daily = (
    df_6m.set_index("timestamp")
        .groupby("type_energie")["conso_clean"]
        .resample("D").sum()
        .reset_index()
)

for e in sorted(daily["type_energie"].unique()):
    tmp = daily[daily["type_energie"] == e]
    ax1.plot(tmp["timestamp"], tmp["conso_clean"], label=e)

ax1.set_title("1) Consommation totale — 6 derniers mois")
ax1.set_xlabel("Date")
ax1.set_ylabel("Conso (kWh ou m³)")
ax1.grid(True, alpha=0.3)
ax1.legend(fontsize=8, title="Énergie")


# Panel 2 

# KPI : conso totale / surface (kWh/m²) sur 6 mois
tmp = df_6m.dropna(subset=["surface_m2"]).copy()
tmp["conso_m2"] = tmp["conso_clean"] / tmp["surface_m2"]
top10 = (
    tmp.groupby(["batiment_id", "nom"], as_index=False)["conso_m2"]
       .sum()
       .sort_values("conso_m2", ascending=False)
       .head(10)
)

ax2.barh(top10["nom"][::-1], top10["conso_m2"][::-1])
ax2.set_title("2) Top 10 bâtiments énergivores (kWh/m² — 6 mois)")
ax2.set_xlabel("Conso cumulée (kWh/m²)")
ax2.grid(True, axis="x", alpha=0.3)

# Panel 3

cost_by_energy = (
    df.dropna(subset=["cout_eur"])
      .groupby("type_energie")["cout_eur"]
      .sum()
      .sort_values(ascending=False)
)

ax3.pie(cost_by_energy.values, labels=cost_by_energy.index, autopct="%1.1f%%")
ax3.set_title("3) Répartition des coûts par énergie")


# Panel 4

order = ["A","B","C","D","E","F","G"]
dpe = df.dropna(subset=["classe_energetique", "conso_par_m2"]).copy()

dpe_stats = (
    dpe.groupby("classe_energetique")["conso_par_m2"]
       .agg(["mean", "std"])
       .reindex(order)
)

ax4.bar(dpe_stats.index, dpe_stats["mean"].values, yerr=dpe_stats["std"].values, capsize=4)
ax4.set_title("4) Conso moyenne par DPE (± écart-type)")
ax4.set_xlabel("Classe DPE")
ax4.set_ylabel("Conso (kWh/m²)")
ax4.grid(True, axis="y", alpha=0.3)


# Panel 5

comm_month = (
    df.groupby(["commune", "month"], as_index=False)["conso_clean"]
      .sum()
)

# Top 12 communes pour lisibilité
top_comm = (
    comm_month.groupby("commune")["conso_clean"].sum()
              .sort_values(ascending=False)
              .head(12)
              .index
)

heat = (
    comm_month[comm_month["commune"].isin(top_comm)]
    .pivot(index="commune", columns="month", values="conso_clean")
)

im = ax5.imshow(heat.values, aspect="auto")
ax5.set_title("5) Heatmap consommation — communes (top 12)")
ax5.set_xlabel("Mois")
ax5.set_ylabel("Commune")

# pour lisibilité
months = heat.columns.to_list()
tick_idx = list(range(0, len(months), max(1, len(months)//6)))
ax5.set_xticks(tick_idx)
ax5.set_xticklabels([months[i].strftime("%Y-%m") for i in tick_idx], rotation=45, ha="right", fontsize=8)

ax5.set_yticks(range(len(heat.index)))
ax5.set_yticklabels(heat.index, fontsize=8)

cbar = fig.colorbar(im, ax=ax5, fraction=0.046, pad=0.04)
cbar.set_label("Conso mensuelle")


# Panel 6 
# conso "cible" : le bâtiment gagne 1 classe DPE (ex: -10% par classe)
# Économie = (conso_actuelle - conso_cible) * tarif moyen énergie

assumption_reduction = 0.10  # -10% par amélioration d'une classe (hypothèse)
# coût total par DPE 
dpe_cost = (
    df.dropna(subset=["classe_energetique", "cout_eur"])
      .groupby("classe_energetique")["cout_eur"]
      .sum()
      .reindex(order)
)

# économie estimée si on réduction 10%
savings = dpe_cost * assumption_reduction

ax6.bar(savings.index, savings.values)
ax6.set_title("6) Économies potentielles si +1 classe DPE (-10% conso)")
ax6.set_xlabel("Classe DPE")
ax6.set_ylabel("Économie estimée (€)")
ax6.grid(True, axis="y", alpha=0.3)

# Finition

fig.suptitle("Dashboard exécutif — Consommations énergétiques bâtiments publics", fontsize=16, y=1.02)
plt.tight_layout()

out = os.path.join(FIG_DIR, "dashboard_energie.png")
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()

print("Export dashboard:", out)
```
