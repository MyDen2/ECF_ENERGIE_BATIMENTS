# Etape 4.1 : Graphiques Matplotlib

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

FIG_DIR = "../output/figures"
os.makedirs(FIG_DIR, exist_ok=True)

df = pd.read_parquet("../output/consommations_enrichies.parquet")

# types
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df = df.dropna(subset=["timestamp", "conso_clean", "type_energie"])
```

## Evolution temporelle de la consommation totale par type d'Ã©nergie (line plot)

```python
df_ts = (
    df.set_index("timestamp")
      .groupby("type_energie")["conso_clean"]
      .resample("D")            # quotidien
      .sum()
      .reset_index()
)

plt.figure(figsize=(12, 6))
for e in sorted(df_ts["type_energie"].unique()):
    tmp = df_ts[df_ts["type_energie"] == e]
    plt.plot(tmp["timestamp"], tmp["conso_clean"], label=e)

plt.title("Ãvolution quotidienne de la consommation totale par type dâÃ©nergie")
plt.xlabel("Date")
plt.ylabel("Consommation (kWh ou mÂ³ selon Ã©nergie)")
plt.legend(title="Type Ã©nergie")
plt.grid(True, alpha=0.3)

out = os.path.join(FIG_DIR, "01_evolution_conso_par_energie.png")
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## Distribution des consommations par type de bÃ¢timent (boxplot)

```python
# boxplote la conso journaliÃ¨re par bÃ¢t.
df_day_build = (
    df.assign(date=df["timestamp"].dt.date)
      .groupby(["type", "batiment_id", "date"], as_index=False)
      .agg(conso_jour=("conso_clean", "sum"))
)

types_order = sorted(df_day_build["type"].dropna().unique())
data = [df_day_build[df_day_build["type"] == t]["conso_jour"].dropna().values for t in types_order]

plt.figure(figsize=(12, 6))
plt.boxplot(data, labels=types_order, showfliers=False)  # sans outliers pour lisibilitÃ©
plt.title("Distribution de la consommation journaliÃ¨re par type de bÃ¢timent")
plt.xlabel("Type de bÃ¢timent")
plt.ylabel("Consommation journaliÃ¨re (kWh ou mÂ³)")
plt.xticks(rotation=30)
plt.grid(True, axis="y", alpha=0.3)

out = os.path.join(FIG_DIR, "02_boxplot_conso_par_type_batiment.png")
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## Heatmap consommation moyenne par heure et jour de semaine

```python
df_hm = df.copy()
df_hm["hour"] = df_hm["timestamp"].dt.hour
df_hm["dow"] = df_hm["timestamp"].dt.dayofweek  # 0=Lun ... 6=Dim

pivot = (
    df_hm.groupby(["dow", "hour"])["conso_clean"]
         .mean()
         .reset_index()
         .pivot(index="dow", columns="hour", values="conso_clean")
)

plt.figure(figsize=(12, 5))
plt.imshow(pivot.values, aspect="auto")
plt.title("Heatmap â Consommation moyenne par heure et jour de semaine")
plt.xlabel("Heure")
plt.ylabel("Jour de semaine (0=Lun ... 6=Dim)")
plt.xticks(range(0, 24, 1), range(0, 24, 1), rotation=0)
plt.yticks(range(0, 7), pivot.index)

cbar = plt.colorbar()
cbar.set_label("Consommation moyenne (kWh ou mÂ³)")

out = os.path.join(FIG_DIR, "03_heatmap_heure_jour.png")
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## Scatter plot tempÃ©rature vs consommation de chauffage avec rÃ©gression

```python
# chauffage ==> gaz
import numpy as np

df_gaz = df[(df["type_energie"] == "gaz") & df["temperature_c"].notna()].copy()

# Option : Ã©chantillonnage si trop gros (garde la forme)
if len(df_gaz) > 200000:
    df_gaz = df_gaz.sample(200000, random_state=42)

x = df_gaz["temperature_c"].values
y = df_gaz["conso_clean"].values

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=5)

# RÃ©gression linÃ©aire
m, b = np.polyfit(x, y, 1)
xx = np.linspace(np.nanmin(x), np.nanmax(x), 200)
yy = m * xx + b
plt.plot(xx, yy)

plt.title("TempÃ©rature vs consommation de gaz (proxy chauffage) + rÃ©gression")
plt.xlabel("TempÃ©rature (Â°C)")
plt.ylabel("Consommation gaz (kWh)")
plt.grid(True, alpha=0.3)

# annotation
corr = np.corrcoef(x, y)[0, 1]
plt.annotate(f"corr={corr:.3f}\ny={m:.2f}x+{b:.2f}", xy=(0.02, 0.95), xycoords="axes fraction",
             ha="left", va="top")

out = os.path.join(FIG_DIR, "04_scatter_temp_vs_gaz_regression.png")
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## Comparaison consommation par classe energÃ©tique (bar chart)

```python
df_dpe = df[df["classe_energetique"].notna() & df["conso_par_m2"].notna()].copy()

order = ["A","B","C","D","E","F","G"]
stats = (
    df_dpe.groupby("classe_energetique")["conso_par_m2"]
          .mean()
          .reindex(order)
)

plt.figure(figsize=(8, 5))
plt.bar(stats.index, stats.values)
plt.title("Consommation moyenne normalisÃ©e par classe Ã©nergÃ©tique (DPE)")
plt.xlabel("Classe Ã©nergÃ©tique (DPE)")
plt.ylabel("Consommation moyenne (kWh/mÂ²)")
plt.grid(True, axis="y", alpha=0.3)

# annotation valeurs
for i, v in enumerate(stats.values):
    if pd.notna(v):
        plt.text(i, v, f"{v:.1f}", ha="center", va="bottom")

out = os.path.join(FIG_DIR, "05_bar_conso_par_classe_dpe.png")
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```
