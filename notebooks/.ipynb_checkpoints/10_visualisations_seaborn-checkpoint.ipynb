# Etape 4.2 : Visualisations Seaborn

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

FIG_DIR = "../output/figures"
os.makedirs(FIG_DIR, exist_ok=True)

df = pd.read_parquet("../output/consommations_enrichies.parquet")
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df = df.dropna(subset=["timestamp", "type_energie", "conso_clean"])

sns.set_theme(style="whitegrid")
```

## Pairplot des consommations (electricite, gaz, eau) par saison

```python
# Cl horaire
df1 = df.copy()
df1["ts_hour"] = df1["timestamp"].dt.floor("H")

# Pivot : une ligne = (batiment_id, ts_hour), colonnes = conso par nrj
wide = (
    df1.pivot_table(index=["batiment_id", "ts_hour"], columns="type_energie", values="conso_clean", aggfunc="sum")
    .reset_index()
)

# saison depuis df1 : la mme pour ts_hour
meta = (
    df1[["batiment_id", "ts_hour", "saison"]]
    .drop_duplicates(subset=["batiment_id", "ts_hour"])
)
wide = wide.merge(meta, on=["batiment_id", "ts_hour"], how="left")

# Garde uniquement les colonnes attendues si prsentes
cols = [c for c in ["electricite", "gaz", "eau"] if c in wide.columns]
wide = wide.dropna(subset=cols + ["saison"])

# petit bout
if len(wide) > 20000:
    wide_plot = wide.sample(20000, random_state=42)
else:
    wide_plot = wide

g = sns.pairplot(wide_plot, vars=cols, hue="saison", corner=True, plot_kws={"s": 8, "alpha": 0.4})
g.fig.suptitle("Pairplot des consommations par saison", y=1.02)

out = os.path.join(FIG_DIR, "06_pairplot_conso_par_saison.png")
g.fig.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## Violin plot de la consommation lectrique par type de btiment

```python
# utilise conso agrge quotidienne par btiment (plus visuel)
df_elec = df[df["type_energie"] == "electricite"].copy()
df_elec["date"] = df_elec["timestamp"].dt.date

elec_day = (
    df_elec.groupby(["type", "batiment_id", "date"], as_index=False)
           .agg(conso_jour=("conso_clean", "sum"))
           .dropna(subset=["type", "conso_jour"])
)

# petit bout
if len(elec_day) > 50000:
    elec_day = elec_day.sample(50000, random_state=42)

plt.figure(figsize=(12, 6))
ax = sns.violinplot(data=elec_day, x="type", y="conso_jour", inner="quartile", cut=0)
ax.set_title("Violin plot  consommation lectrique journalire par type de btiment")
ax.set_xlabel("Type de btiment")
ax.set_ylabel("Consommation lectrique journalire (kWh)")
plt.xticks(rotation=30)

out = os.path.join(FIG_DIR, "07_violin_electricite_par_type_batiment.png")
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## Heatmap annote de la matrice de corrlation complte

```python
corr_cols = [
    "conso_clean",
    "temperature_c", "humidite_pct", "rayonnement_solaire_wm2", "vitesse_vent_kmh", "precipitation_mm",
    "surface_m2", "nb_occupants_moyen", "annee_construction",
    "conso_par_m2", "conso_par_occupant", "cout_eur", "IPE_kwh_m2_an"
]
corr_cols = [c for c in corr_cols if c in df.columns]

corr_df = df[corr_cols].dropna()
# petit bout car norme
if len(corr_df) > 200000:
    corr_df = corr_df.sample(200000, random_state=42)

corr = corr_df.corr(numeric_only=True)

plt.figure(figsize=(12, 10))
ax = sns.heatmap(corr, annot=True, fmt=".2f")
ax.set_title("Heatmap de la matrice de corrlation (variables cls)")

out = os.path.join(FIG_DIR, "08_heatmap_correlation_annotee.png")
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## FacetGrid : volution mensuelle par commune (top 6 communes)

```python
import matplotlib.dates as mdates

df4 = df.copy()

# Cl temporelle mensuelle (datetime rel)
df4["month"] = df4["timestamp"].dt.to_period("M").dt.to_timestamp()

top_communes = (
    df4.groupby("commune")["conso_clean"].sum()
       .sort_values(ascending=False)
       .head(6)
       .index
)

monthly_city = (
    df4[df4["commune"].isin(top_communes)]
    .groupby(["commune", "month"], as_index=False)
    .agg(conso_mois=("conso_clean", "sum"))
    .sort_values(["commune", "month"])
)

g = sns.FacetGrid(
    monthly_city,
    col="commune",
    col_wrap=3,        # 2 lignes  3 colonnes
    sharey=False,
    height=3.5,        
    aspect=1.4         
)

g.map_dataframe(sns.lineplot, x="month", y="conso_mois")
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Mois", "Consommation mensuelle")

# Gestion des dates
for ax in g.axes.flatten():
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
    for label in ax.get_xticklabels():
        label.set_rotation(45)
        label.set_ha("right")

g.fig.subplots_adjust(top=0.88, hspace=0.35)
g.fig.suptitle("volution mensuelle de la consommation  Top 6 communes")

out = os.path.join(FIG_DIR, "09_facetgrid_evolution_mensuelle_top6_communes.png")
g.fig.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## Jointplot : relation surface vs consommation avec distributions marginales

```python
#conso journalire ou mensuelle par btiment pour lisibilit
df5 = df.dropna(subset=["surface_m2"]).copy()
df5["date"] = df5["timestamp"].dt.date

day_build = (
    df5.groupby(["batiment_id", "type", "surface_m2", "date"], as_index=False)
       .agg(conso_jour=("conso_clean", "sum"))
)

# petit bout
if len(day_build) > 30000:
    day_build_plot = day_build.sample(30000, random_state=42)
else:
    day_build_plot = day_build

g = sns.jointplot(data=day_build_plot, x="surface_m2", y="conso_jour", kind="scatter", height=7)
g.fig.suptitle("Relation surface vs consommation journalire (avec distributions)", y=1.02)
g.set_axis_labels("Surface (m)", "Consommation journalire (kWh ou m)")

out = os.path.join(FIG_DIR, "10_jointplot_surface_vs_conso.png")
g.fig.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```

## Catplot : consommation par classe nergetique et type de btiment

```python
# prend une mesure normalise (conso_par_m2) pour viter les pbs de taille
df6 = df.dropna(subset=["classe_energetique", "type", "conso_par_m2"]).copy()

# petit bout
if len(df6) > 80000:
    df6 = df6.sample(80000, random_state=42)

order_dpe = ["A","B","C","D","E","F","G"]

g = sns.catplot(
    data=df6,
    x="classe_energetique", y="conso_par_m2",
    hue="type",
    kind="bar",
    order=order_dpe,
    height=6, aspect=1.6
)
g.fig.suptitle("Consommation normalise (kWh/m) par classe DPE et type de btiment", y=1.02)
g.set_axis_labels("Classe nergtique (DPE)", "Consommation (kWh/m)")

out = os.path.join(FIG_DIR, "11_catplot_dpe_vs_type_conso_par_m2.png")
g.fig.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
```
