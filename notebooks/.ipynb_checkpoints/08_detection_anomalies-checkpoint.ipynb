# Etape 3.3 : Détection d'anomalies

```python
import pandas as pd

df = pd.read_parquet("../output/consommations_enrichies.parquet")
df
```

## Identifier les pics de consommation anormaux (>3 écarts-types)

```python
stats_energy = (
    df.groupby("type_energie")["conso_clean"]
    .agg(["mean", "std"])
    .reset_index()
)

df = df.merge(stats_energy, on="type_energie", how="left")

df["pic_anormal"] = df["conso_clean"] > (df["mean"] + 3 * df["std"])
df
```

## Détecter les périodes de sous-consommation suspectes (bâtiment ferme non declaré)

```python
seuil_bas = (
    df.groupby("type_energie")["conso_clean"]
    .quantile(0.05)
    .reset_index()
    .rename(columns={"conso_clean": "seuil_bas"})
)

df = df.merge(seuil_bas, on="type_energie", how="left")

df["sous_conso_suspecte"] = df["conso_clean"] < df["seuil_bas"]
df
```

## Repérer les bâtiments dont la consommation ne correspond pas à leur DPE

```python
# choix du seuil de 200 kWh/m²/an
df["dpe_incoherent"] = (
    df["classe_energetique"].isin(["A", "B", "C"]) &
    (df["IPE_kwh_m2_an"] > 200)
)
df
```

## Lister les bâtiments nécessitant un audit énergétique

```python
df["audit_recommande"] = (
    df["pic_anormal"] |
    df["sous_conso_suspecte"] |
    df["dpe_incoherent"] |
    (df["ecart_moyenne_categorie"].abs() > 50)
)
df
```

## Export anomalies_détctées

```python
df_anomalies = df[df["audit_recommande"]].copy()

df_anomalies_export = df_anomalies[[
    "batiment_id",
    "nom",
    "type",
    "commune",
    "type_energie",
    "conso_clean",
    "IPE_kwh_m2_an",
    "classe_energetique",
    "pic_anormal",
    "sous_conso_suspecte",
    "dpe_incoherent",
    "ecart_moyenne_categorie"
]]

df_anomalies_export.to_csv(
    "../output/anomalies_detectees.csv",
    index=False
)

df_anomalies_export.shape
```

## Rapport de recommandation

# Rapport de recommandations d’audit énergétique

## Contexte
Une analyse des consommations énergétiques a été menée afin d’identifier
les bâtiments présentant des comportements atypiques ou incohérents.

## Anomalies détectées
Les anomalies suivantes ont été mises en évidence :
- Pics de consommation dépassant significativement les valeurs habituelles
- Périodes prolongées de sous-consommation suggérant une fermeture non déclarée
- Incohérences entre la classe énergétique (DPE) et la consommation réelle
- Écarts importants par rapport à la moyenne des bâtiments de même catégorie

## Bâtiments à auditer en priorité
Les bâtiments cumulant plusieurs signaux d’alerte sont considérés comme
prioritaires pour un audit énergétique approfondi.

## Recommandations
- Vérifier l’état des équipements (chauffage, climatisation, éclairage)
- Contrôler la qualité et la fiabilité des capteurs de mesure
- Analyser les usages réels et les plages d’occupation
- Mettre à jour les informations DPE si nécessaire

## Conclusion
La détection précoce de ces anomalies permet d’optimiser les actions
d’amélioration énergétique et de réduire les coûts d’exploitation.
